```python
#Reshaping numpy array
mat=np.array([[10,20,30],[40,50,60]])
print(mat)
```

```python
mat=np.array([[10,20,30],[40,50,60]])
mat_reshaped=mat.reshape(3,2)
flat=mat.ravel()
print('original\n',mat)
print('reshaped\n',mat_reshaped)
print('flattened array\n',flat)
```

```python
mat1=np.array([[[10,20,30],[40,50,60]],[[3,2,1],[4,5,6]]])
mat1_reshaped=mat1.reshape(2,2,3)
print('original\n',mat1)
print('reshaped\n',mat1_reshaped)
```
