```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math
```

```python
np.random.seed(4)
```

```python
p=0.3
n_samples=10000
bern=np.random.binomial(1,p,size=n_samples)
print("Bernouli sample mean: ",bern.mean())
print("Bernouli sample val: ",bern.var())
```

```python
values,counts = np.unique(bern,return_counts=True)
print(values)
print(counts)
```

```python
plt.bar(values,counts/len(bern))
plt.title("Bernouli PMF")
plt.xlabel("values")
plt.ylabel("Probability")
```

```python
#BINOMIAL
n,p=10,0.4
k=np.arange(0,n-1)
pmf_binom=stats.binom.pmf(k,n,p)
print("Binomial theoretical mean: ", stats.binom.mean(n,p))
print("Binomial theoretical var: ", stats.binom.var(n,p))
```

```python
#POISSON
lam=3
k_p=np.arange(0,15)
pmf_poisson=stats.poisson.pmf(k_p,lam)
print("Binomial theoretical mean: ", stats.poisson.mean(lam))
print("Binomial theoretical var: ", stats.poisson.var(lam))
```

```python

fig,ax=plt.subplots(1,2,figsize=[10,4])
ax[0].stem(k,pmf_binom)
ax[0].set_title("binomial PMF: ")
ax[1].stem(k_p,pmf_poisson)
ax[1].set_title("poisson PMF: ")
plt.tight_layout()
```

```python
p = 0.4
n = 1   
k = np.arange(0, 2)  
pmf_bernoulli = stats.bernoulli.pmf(k, p)
print("Bernoulli theoretical mean: ",stats.bernoulli.mean(p))
print("Bernoulli theoretical var: ",stats.bernoulli.var(p))
```

```python
#Normal(continuous) example
mew, sigma=2.0, 1.5
norm_samples = np.random.normal(mew, sigma, size=n_samples)
print('Normal sample mean(approx mu):',norm_samples.mean())
print('Normal sample mean(approx sigma^2):',norm_samples.var())
```

```python
#Normal PDF (empirical histogram+theoretical PDF)
plt.hist(norm_samples,bins=40,density=True)
x=np.linspace(mew-4*sigma,mew+4*sigma,200)
plt.plot(x,stats.norm.pdf(x,mew,sigma))
plt.title("Normal PDF(empirical+theoretical)")
plt.tight_layout()
```

```python
# Uniform(a, b)
a, b = 0, 5
uni_samps = np.random.uniform(a, b, size=10000)
x = np.linspace(a, b, 200)

# Exponential
lam = 1.2
exp_samps = np.random.exponential(1/lam, size=10000)
x_exp = np.linspace(0, 8, 200)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].hist(uni_samps, bins=40, density=True)
ax[0].plot(x, stats.uniform.pdf(x, a, b-a))
ax[0].set_title('Uniform distribution (empirical + theoretical)')

ax[1].hist(exp_samps, bins=40, density=True)
ax[1].plot(x_exp, stats.expon.pdf(x_exp, scale=1/lam))
ax[1].set_title('Exponential distribution (empirical + theoretical)')

plt.tight_layout()
plt.show()

# Normal shown previously; we can show standard normal CDF example
xs = np.linspace(-4, 4, 200)
plt.plot(xs, stats.norm.cdf(xs))
plt.title('Standard Normal CDF')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate correlated variables
mean = [0, 0]
cov = [[1.0, 0.6], 
       [0.6, 1.0]]   # covariance matrix with positive correlation

samples = np.random.multivariate_normal(mean, cov, size=5000)
x = samples[:, 0]
y = samples[:, 1]

print('Empirical covariance:', np.cov(x, y)[0, 1])
print('Empirical correlation:', np.corrcoef(x, y)[0, 1])

# Scatter plot
plt.figure(figsize=(6, 5))
plt.scatter(x[:500], y[:500], s=10, alpha=0.6)
plt.title('Scatter plot of correlated variables')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
