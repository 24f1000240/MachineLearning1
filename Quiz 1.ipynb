```python
marks = int(input('''Enter marks'(0-100)': '''))
if 0<=marks<=100:
    if marks>=90: print("A")
    elif 75<=marks<=89: print("B")
    elif 50<=marks<=74: print("C")
    else: print("Failed")
else: print("Invalid Range")
```

```python
def isprime(n):
    if n<=1:
        return False
    if n==2:
        return True
    for i in range(2,n):
        if n%i==0:
            return False
        else: return True
```

```python
while True:
    n=(input("Enter digit to check for prime: "))
    if n =="break":
        break
    else: print(isprime(int(n)))
```
