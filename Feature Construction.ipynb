```python
import pandas as pd
import numpy as np
```

```python
room_length=[10,20,12,34,43,20]
room_breadth=[20,20,10,11,25,16]
room_type=["big","big","normal","big","big","normal"]
data=pd.DataFrame({"length":room_length,
                  "breadth":room_breadth,
                   "type":room_type
                  })
data
```

```python
data["Area"]=data["length"]*data["breadth"]
```

```python
data
```

```python
#Encoding categorical(Nominal)variables->One-Hot Encoding
age=[18,20,21,34,56,57,45]
city=["City A","City B","City A","City A","City C","City B","City A"]
data1=pd.DataFrame({"Age":age,"City":city})
print(data1)
dummy_features= pd.get_dummies(data1["City"])
print(dummy_features)
data_age=pd.DataFrame(data=data1,columns=["Age"])
data_mod=pd.concat([data_age,dummy_features],axis=1)
print("Final Dataset \n",data_mod)
```

```python
#encoding categorical (ordinal) variables
marks_science=[78,50,87,91,45,62]
marks_maths=[75,62,98,95,42,57]
grade=['B','C','A','A','D','B']
data=pd.DataFrame({'Science Marks':marks_science,
                  'Maths Marks': marks_maths,
                  'Total Grade':grade})
print(data)
```

```python
#option1
from  sklearn import preprocessing
le=preprocessing.LabelEncoder()
le.fit(data['Total Grade'])
data['Total Grade']=le.transform(data['Total Grade'])
print(data)
```

```python
#option2
target=data['Total Grade'].replace(['A','B','C','D'],[0,1,2,3])
predictors=data.iloc[:,0:2]
data_mod=pd.concat([predictors,target],axis=1)
data_mod
```
