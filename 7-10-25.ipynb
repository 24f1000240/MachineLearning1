```python
df=pd.DataFrame({"Amit":[10,20,30],
                 "Barsa":[40,50,60],
                 "Chinmay":[70,45,90]}, index=["x","y","z"]
                )
df
```

```python
#display pass student marks>30
print("pass student marks>30")
print(df[df>=30])
```

```python
#display pass student marks>30
print("pass student marks>30 and 70")
print(df[(df>=30) & (df<=70)])
```
